Extraction de la raison principale des appels

```python
def createTitle(row):
    a = row.split(":")[0]
    return a
```

```python
df["raison"]=df['title'].apply(lambda x: createTitle(x))
```

```python
 df["raison"]=df['title'].apply(lambda x: x.split(":")[0])
```

```python
# Plus lisible, plus rapide (vectorisé, pas besoin de
df['raison'] = df['title'].str.split(':').str[0]
```

 On ajoute les colonnes Heure, Jour, Mois, Jour_semaine

```python
df["Heure"] = df["dateTime"].dt.hour
df["Jour"] = df["dateTime"].dt.day
df["Mois"] = df["dateTime"].dt.month
df["Jour_semaine"] = df["dateTime"].dt.day_name()
# on cree une variable qui contient le jour de la semaine
df
```

On afficher la répartition des appels par jour de la semaine

Le mois qui a eu le plus d'appels

```python
df['Mois'].value_counts().sort_index()
```

```python
# Créer un dictionnaire pour mapper les numéros des mois à leurs noms
mois_noms = {
    1: "Janvier", 2: "Février", 3: "Mars", 4: "Avril", 5: "Mai", 6: "Juin",
    7: "Juillet", 8: "Août", 9: "Septembre", 10: "Octobre", 11: "Novembre", 12: "Décembre"
}

# Compter les appels par mois et mapper les numéros aux noms
appels_par_mois = df['Mois'].value_counts().sort_index()
appels_par_mois.index = appels_par_mois.index.map(mois_noms)

# Afficher les résultats
print(appels_par_mois)
```

On crée une heatmap du nombre d'appels par jour de la semaine et heure , on verra les heures de la journée où il y a le plus d’appels, par jour.

```python
heat_data = df.groupby(['Jour_semaine', 'Heure']).count()['title'].unstack()
sns.heatmap(heat_data, cmap='viridis')
plt.title("Appels par heure et jour")
plt.xlabel("Heure")
plt.ylabel("Jour de la semaine")    


```

```python

sns.set_theme(style="white", context="talk")

# Préparer les données pour la heatmap
heat_data = df.groupby(['Jour_semaine', 'Heure']).count()['title'].unstack()

# Réorganiser les jours de la semaine dans l'ordre souhaité
heat_data = heat_data.reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])

# Configurer la figure matplotlib
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12), sharex=True)

# Heatmap 1 : Viridis
sns.barplot(heat_data, cmap='viridis', cbar_kws={'label': 'Nombre d\'appels'}, ax=ax1)
ax1.set_title("Heatmap avec Viridis")
ax1.set_xlabel("Heure")
ax1.set_ylabel("Jour de la semaine")

# Heatmap 2 : Coolwarm
sns.barplot(heat_data, cmap='coolwarm', cbar_kws={'label': 'Nombre d\'appels'}, ax=ax2)
ax2.set_title("Heatmap avec Coolwarm")
ax2.set_xlabel("Heure")
ax2.set_ylabel("Jour de la semaine")

# Heatmap 3 : Plasma
sns.barplot(heat_data, cmap='plasma', cbar_kws={'label': 'Nombre d\'appels'}, ax=ax3)
ax3.set_title("Heatmap avec Plasma")
ax3.set_xlabel("Heure")
ax3.set_ylabel("Jour de la semaine")

# Finaliser la mise en page
sns.despine(bottom=True)
plt.setp(f.axes, yticks=[])
plt.tight_layout(h_pad=2)
plt.show()
```
