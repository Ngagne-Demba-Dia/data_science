Diagramme en barres pour comparer le taux de survie entre hommes et femmes
hue pour separer les hommes et femmes qui on survecu et non survecu


 Diagramme en violon pour comparer l'âge des survivants et des non-survivants

```python
sns.violinplot(x='Survived', y='Age', data=df)
plt.title('Survived vs. Age')
plt.show()
```

 Diagramme en boîte pour comparer la classe des survivants et des non-survivants

```python

sns.boxplot(x='Pclass', y='Survived', data=df)
plt.title('Survived vs. Pclass')
plt.show()
```

```python
survived = df["Survived"].value_counts()
survived.plot(kind="bar",color=["red","green"])
plt.title("survivants")
plt.ylabel("nombres de passages")
plt.xticks()
plt.show()
```

Y a-t-il une interaction entre classe (Pclass), sexe et survie ?

```python
sns.catplot(x="Pclass", hue="Survived", col="Sex", data=df, kind="count")
```

On visualise l’âge selon classe et survie 

```python
sns.violinplot(x="Pclass", y="Age", hue="Survived", data=df, split=True)
```

On crée des catégories d’âge (enfant, jeune, adulte, senior) et voir qui a survécu

```python
def categoriser_age(age):
    if age < 12:
        return "Enfant"
    elif age < 25:
        return "Jeune"
    elif age < 60:
        return "Adulte"
    else:
        return "Senior"

df['age_cat'] = df['Age'].apply(lambda x: categoriser_age(x) if pd.notnull(x) else "Inconnu")
sns.countplot(x='age_cat', hue='Survived', data=df)
plt.title("Survie par catégorie d'âge")
plt.show()
```
